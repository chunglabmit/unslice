```python
def get_orientation(pts, voxel_size=(1,1,1)):
    direction = [pts[-1][k]*voxel_size[k] - pts[0][k]*voxel_size[k] for k in range(3)]
    r = np.linalg.norm(direction)
    phi = np.arctan2(direction[1],direction[0])*180/np.pi
    theta = np.arccos(direction[2]/r)*180/np.pi
    return phi,theta,r
```

```python
f = open(bdir('3-fibers_LL.json'))
data = json.load(f)
```

```python
axons = []
lengths = []
orientations = []
voxel_size = (.306,.433,.306)
for dataset in tqdm(data):
    axon = np.array(dataset['points'])
    # compute orientation
    phi,theta,r = get_orientation(axon,voxel_size)
    
    # compute length
    length = 0
    for idx,pt in enumerate(axon[:-1]):
        length += np.linalg.norm([(axon[idx+1][k]-pt[k])*voxel_size[k] for k in range(3)])
    
    if r>0:
        orientations.append([phi,theta,r])
        lengths.append(length)
        axons.append(axon)
    
    
    
```

```python
# Convert orientations to histogram

# bins go phi = 0-45(ENE),45-90(NNE),90-135(NNW),135-180(WNW),-180--135(WSW),-135--90(SSW),-90--45(SSE),-45-0(ESE)
bins_theta0_45 = [0]*8 # less vertical fibers
bins_theta45_90 = [0]*8 # more vertical fibers 

for orientation in tqdm(orientations_):
    phi,theta,_ = orientation
    if (theta<90) and (theta>=45):
        if (phi>=0) and (phi<45):
            bins_theta0_45[0] += 1
        elif (phi>=45) and (phi<90):
            bins_theta0_45[1] += 1
        elif (phi>=90) and (phi<135):
            bins_theta0_45[2] += 1
        elif (phi>=135) and (phi<180):
            bins_theta0_45[3] += 1
        elif (phi>=-180) and (phi<-135):
            bins_theta0_45[4] += 1
        elif (phi>=-135) and (phi<-90):
            bins_theta0_45[5] += 1
        elif (phi>=-90) and (phi<-45):
            bins_theta0_45[6] += 1
        elif (phi>=-45) and (phi<0):
            bins_theta0_45[7] += 1
    elif (theta<135) and (theta>=90):
        if (phi>=0) and (phi<45):
            bins_theta0_45[4] += 1
        elif (phi>=45) and (phi<90):
            bins_theta0_45[5] += 1
        elif (phi>=90) and (phi<135):
            bins_theta0_45[6] += 1
        elif (phi>=135) and (phi<180):
            bins_theta0_45[7] += 1
        elif (phi>=-180) and (phi<-135):
            bins_theta0_45[0] += 1
        elif (phi>=-135) and (phi<-90):
            bins_theta0_45[1] += 1
        elif (phi>=-90) and (phi<-45):
            bins_theta0_45[2] += 1
        elif (phi>=-45) and (phi<0):
            bins_theta0_45[3] += 1
            
    elif (theta<45) and (theta>=0):
        if (phi>=0) and (phi<45):
            bins_theta45_90[0] += 1
        elif (phi>=45) and (phi<90):
            bins_theta45_90[1] += 1
        elif (phi>=90) and (phi<135):
            bins_theta45_90[2] += 1
        elif (phi>=135) and (phi<180):
            bins_theta45_90[3] += 1
        elif (phi>=-180) and (phi<-135):
            bins_theta45_90[4] += 1
        elif (phi>=-135) and (phi<-90):
            bins_theta45_90[5] += 1
        elif (phi>=-90) and (phi<-45):
            bins_theta45_90[6] += 1
        elif (phi>=-45) and (phi<0):
            bins_theta45_90[7] += 1
            
    elif (theta<180) and (theta>=135):
        if (phi>=0) and (phi<45):
            bins_theta45_90[4] += 1
        elif (phi>=45) and (phi<90):
            bins_theta45_90[5] += 1
        elif (phi>=90) and (phi<135):
            bins_theta45_90[6] += 1
        elif (phi>=135) and (phi<180):
            bins_theta45_90[7] += 1
        elif (phi>=-180) and (phi<-135):
            bins_theta45_90[0] += 1
        elif (phi>=-135) and (phi<-90):
            bins_theta45_90[1] += 1
        elif (phi>=-90) and (phi<-45):
            bins_theta45_90[2] += 1
        elif (phi>=-45) and (phi<0):
            bins_theta45_90[3] += 1
```

```python
# convert thetas to 0 to 90
print(orientations_[:,1].max())
phis = orientations_[:,0].copy()
thetas = orientations_[:,1].copy()

phis[(thetas>90)*(phis<0)] += 180
phis[(thetas>90)*(phis>=0)] -= 180
thetas[thetas>90] = 180-thetas[thetas>90]
print(phis.max(),phis.min(),thetas.max())
```

```python
# Convert orientations to histogram
bin_size = 30
bin_angles = [(i,j) for j in np.arange(0,90,bin_size) for i in np.arange(-180,180,bin_size) ]
counts = []

for idx,bin_angle in enumerate(bin_angles):
    phi0,theta0 = bin_angle
#     print(theta0,phi0)
    if phi0 >= 180-bin_size:
        phi1 = 1000
    else:
        phi1 = phi0+bin_size
    if theta0 >= 90-bin_size:
        theta1 = 1000
    else:
        theta1 = theta0+bin_size
#     print(phi1,theta1)
# #     print(phi0,theta0,phi1,theta1)
#     print(np.sum((phis>=phi0)))
#     print(np.sum((phis<phi1)))
#     print(np.sum((thetas>=theta0)))
#     print(np.sum((thetas<theta1)))
    counts.append(np.sum((phis>=phi0)*(phis<phi1)*(thetas>=theta0)*(thetas<theta1)))
    
        
```
